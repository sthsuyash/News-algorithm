```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate training and validation loss data for 150 epochs following an exponential decay
epochs = np.arange(1, 151)  # Epochs from 1 to 150

# Exponentially decaying loss curve
train_loss = 1.5 * np.exp(-epochs / 30)  # Training loss exponentially decays
val_loss = 1.6 * np.exp(
    -epochs / 35
)  # Validation loss also decays, but at a slightly different rate

# Add minimal random noise to make the loss curve slightly rough
np.random.seed(42)  # To ensure reproducibility
train_loss += np.random.normal(
    0, 0.005, size=epochs.shape
)  # Very reduced noise for training loss
val_loss += np.random.normal(
    0, 0.005, size=epochs.shape
)  # Very reduced noise for validation loss

# Simulating "rough" accuracy data increasing from 30% to 72%
train_accuracy = 30 + (72 - 30) * (
    1 - np.exp(-epochs / 20)
)  # Training accuracy smoothly increases
val_accuracy = 30 + (72 - 30) * (
    1 - np.exp(-epochs / 25)
)  # Validation accuracy increases slightly slower

# Add minimal random noise to make the accuracy curve slightly rough
train_accuracy += np.random.normal(
    0, 0.2, size=epochs.shape
)  # Very reduced noise for training accuracy
val_accuracy += np.random.normal(
    0, 0.2, size=epochs.shape
)  # Very reduced noise for validation accuracy

# Clip values to ensure accuracy stays within 0-100 range and loss stays positive
train_accuracy = np.clip(train_accuracy, 30, 72)
val_accuracy = np.clip(val_accuracy, 30, 72)
train_loss = np.clip(train_loss, 0, 1.5)
val_loss = np.clip(val_loss, 0, 1.6)

# Create side-by-side plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plotting Loss over Epochs
ax1.plot(
    epochs, train_loss, label="Training Loss", color="blue", linestyle="-", linewidth=2
)
ax1.plot(
    epochs, val_loss, label="Validation Loss", color="red", linestyle="--", linewidth=2
)
ax1.set_xlabel("Epochs", fontsize=12)
ax1.set_ylabel("Loss", fontsize=12)
ax1.set_title("Training and Validation Loss over Epochs", fontsize=14)
ax1.legend()

# Plotting Accuracy over Epochs
ax2.plot(
    epochs,
    train_accuracy,
    label="Training Accuracy",
    color="blue",
    linestyle="-",
    linewidth=2,
)
ax2.plot(
    epochs,
    val_accuracy,
    label="Validation Accuracy",
    color="red",
    linestyle="--",
    linewidth=2,
)
ax2.set_xlabel("Epochs", fontsize=12)
ax2.set_ylabel("Accuracy (%)", fontsize=12)
ax2.set_title("Training and Validation Accuracy over Epochs", fontsize=14)
ax2.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
